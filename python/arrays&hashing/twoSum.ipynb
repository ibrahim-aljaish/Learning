# two sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

### example
Input: nums = [2,7,11,15], target = 9 <br>
Output: [0,1] <br>
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

# Brute forcing


we take first number in our list and check its sum with all other entries in the list. it it matches the target value.
repeat with the second number and so on until the traget is found

````python
def twoSumBruteForce( nums:list[int], target: int)-> list[int]:
    sum = 0;
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i] + nums[j] == target:
                return [i,j]
    return []


````

# using a hashmap



````python

def twoSumOnePass(self, nums: list[int], target: int) -> list[int]:
    prevMap = {} # val : index

    for i,n in enumerate(nums):
        diff = target - n
        if diff in prevMap:
            return [prevMap[diff],i]
        prevMap[n] = i
    return
````

### why was this failing the second test?

Your twoSumBruteForce function fails the second test case because it returns an incorrect output for the input [3, 2, 4], 6. Specifically, it returns [0, 0], which is not a valid solution.

The problem with your implementation is that the inner loop iterates over all indices of the input list, including the current index i. This means that your function considers pairs of the form (i, i), which are not valid solutions (unless the same element appears twice in the input list).

To fix this issue, you should modify your inner loop to start from i+1 instead of 0. This ensures that the inner loop only considers indices that come after the current index i.
